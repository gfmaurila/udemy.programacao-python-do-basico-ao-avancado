# 游 Mini Sistema de Cadastro Interativo com Salvar, Carregar e Analisar

```

import csv
import os
import pandas as pd

arquivo_csv = "cadastros.csv"

def cadastrar():
    entrada = input("Digite nome, idade, cidade e sal치rio separados por v칤rgula: ")
    try:
        nome, idade, cidade, salario = entrada.split(",")
        idade = int(idade.strip())
        salario = float(salario.strip().replace(",", "."))
        with open(arquivo_csv, mode="a", newline='', encoding="utf-8") as f:
            escritor = csv.writer(f)
            escritor.writerow([nome.strip(), idade, cidade.strip(), salario])
        print("Cadastro salvo com sucesso!")
    except ValueError:
        print("Erro: dados inv치lidos. Use o formato: nome,idade,cidade,sal치rio")

def mostrar_dados():
    if not os.path.exists(arquivo_csv):
        print("Nenhum dado cadastrado ainda.")
        return
    df = pd.read_csv(arquivo_csv, names=["Nome", "Idade", "Cidade", "Sal치rio"])
    display(df)

def analisar_dados():
    if not os.path.exists(arquivo_csv):
        print("Nenhum dado para an치lise.")
        return
    df = pd.read_csv(arquivo_csv, names=["Nome", "Idade", "Cidade", "Sal치rio"])
    print("M칠dia salarial:", df["Sal치rio"].mean())
    print("Pessoa mais jovem:", df.loc[df["Idade"].idxmin()]["Nome"])
    print("Pessoa mais velha:", df.loc[df["Idade"].idxmax()]["Nome"])
```

```

# Menu interativo
while True:
    print("\n--- MENU ---")
    print("1 - Cadastrar pessoa")
    print("2 - Mostrar todos os cadastros")
    print("3 - Analisar dados")
    print("4 - Sair")

    opcao = input("Escolha uma op칞칚o: ")

    if opcao == "1":
        cadastrar()
    elif opcao == "2":
        mostrar_dados()
    elif opcao == "3":
        analisar_dados()
    elif opcao == "4":
        print("Encerrando...")
        break
    else:
        print("Op칞칚o inv치lida.")
```
